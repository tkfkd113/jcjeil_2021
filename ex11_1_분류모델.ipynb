# k-Nearest Neighborhood(k-NN)

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
breast_cancer_data = load_breast_cancer()
```

```python
df_data = pd.DataFrame(breast_cancer_data.data)
df_labels = pd.DataFrame(breast_cancer_data.target)
```

```python
from sklearn.model_selection import train_test_split

training_data, validation_data, training_labels, valiaation_labels = train_test_split(df_data, df_labels, test_size=0.2, random_state = 100)

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=3)
classifier

classifier.fit(training_data, training_labels)
print(classifier.score(validation_data, valiaation_labels))
```

```python
import matplotlib.pyplot as plt
k_list = range(1,101)
accuracies = []

for k in k_list:
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(training_data, training_labels)
  accuracies.append(classifier.score(validation_data, valiaation_labels))

plt.plot(k_list, accuracies)
plt.xlabel('k')
plt.ylabel('Validation Accuracy')
plt.title('Breast Cancer Classifier Accuracy')
plt.show()
```

#선형 분류 보델

```python
from sklearn.svm import LinearSVC
import numpy as np
import mglearn

linear_svm = LinearSVC(C=0.01)
linear_svm.fit(X, y)

mglearn.plots.plot_2d_classification(linear_svm, X, fill=True, alpha=0.7)
mglearn.discrete_scatter(X[:,0], X[:,1], y)
line = np.linspace(-15,15)

for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, mglearn.cm3.colors):
  plt.plot(line, -(line * coef[0]+intercept) / coef[1], c=color)
  plt.legend(['class 0', 'class 1', 'class 2','class 0 w', 'class 1 w', 'class 2 w'])
  plt.xlabel('특성 0')
  plt.ylabel('특성 1')

plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print(cancer['DESCR'])
```

#선형 분류 모델 - Logistic Regression

#Logistic Regression으로 분류-기본값 C =1




```python
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=0)

logreg = LogisticRegression().fit(X_train, y_train)
print('훈련 세트 점수: {:.3f}'.format(logreg.score(X_train, y_train)))
print('테스트 세트 점수: {:.3f}'.format(logreg.score(X_test, y_test)))
```

#Logistic Regression으로 분류-기본값 C =100

```python
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=0)

logreg100 = LogisticRegression(C=100).fit(X_train, y_train)
print('훈련 세트 점수: {:.3f}'.format(logreg100.score(X_train, y_train)))
print('테스트 세트 점수: {:.3f}'.format(logreg100.score(X_test, y_test)))
```

Logistic Regression으로 분류-기본값 C =0.01

```python
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=0)

logreg001 = LogisticRegression(C=0.01).fit(X_train, y_train)
print('훈련 세트 점수: {:.3f}'.format(logreg001.score(X_train, y_train)))
print('테스트 세트 점수: {:.3f}'.format(logreg001.score(X_test, y_test)))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.plot(logreg.coef_.T, 'o', label="C=1")
plt.plot(logreg100.coef_.T, '^', label="C=100")
plt.plot(logreg001.coef_.T, 'v', label="C=0.01")

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-5,5)
plt.xlabel('특성')
plt.ylabel('계수의 크기')
plt.legend()
plt.show()
```
